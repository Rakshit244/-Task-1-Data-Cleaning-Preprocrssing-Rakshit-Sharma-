Project Objective:

Clean and Prepare raw data for analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
df.rename(columns ={
    'PassengerId': 'Passenger_Id',
    'PClass': 'Passenger_Class',
    'Sex': 'Gender'
}, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv("Cleaned_titanic_dataset.csv", index = False)
```

Key Indicators:

- Data types were refined to improve analytical accuracy.

- The Age column had multiple null entries.

- A vast majority of the Cabin entries were missing.

- All duplicate records were removed.

- The resultent dataset is more refined and consistent making it reay for furthur analysis.
